```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
import matplotlib
import numpy as np 
from google.colab import files
```

```python
rainAUS_df=pd.read_csv('/content/drive/MyDrive/archive (4)/weatherAUS.csv')
rainAUS_df.dropna(inplace=True)
rainAUS_df
```

```python
# Visu 1 heatmap
```

```python
matrix=pd.pivot_table(data=rainAUS_df,values='Rainfall',index='WindGustDir',columns='Cloud9am',aggfunc='mean')
plt.figure(dpi=150)
plt.title('Milimetros promedio segun direccion del viento y nubosidad a las 9 a.m ')
fig = sns.heatmap(matrix,vmin=0,vmax=8,cmap='Blues')
plt.xlabel("Indice de nubosidad")
plt.ylabel("Direccion del viento")
plt.savefig('visu.jpg', dpi=200, bbox_inches='tight', pad_inches=0.5)
```

```python
# visu 2 barplot
```

```python
plt.figure(dpi=150)
sns.barplot(data=rainAUS_df,x='Location',y='Rainfall',color='lightblue',ci=None)
plt.ylim([0,5])
plt.locator_params(axis="y", nbins=10)
plt.xticks(rotation=85)
plt.title('Milimetros por localidad')
plt.xlabel("Localidad")
plt.ylabel("Mm de lluvia")
```

```python
flights_df=pd.read_csv('/content/flight_delays_train.csv')
```

```python
df_aux = flights_df.loc[flights_df['dep_delayed_15min']== 'Y']
df_aux['dep_delayed_15min'].replace('Y',1,inplace=True)

```

```python
df_aux['DayOfWeek'].replace(['c-1','c-2','c-3','c-4','c-5','c-6','c-7'],['lunes','martes','miercoles','jueves','viernes','sabado','domingo'],inplace=True)
```

```python
last_df = df_aux.groupby(['DayOfWeek']).agg({'dep_delayed_15min':['sum']})
level0=last_df.columns.get_level_values(0)
level1=last_df.columns.get_level_values(1)
last_df.columns = level0 + '_' + level1
last_df.reset_index(inplace=True)
df_aux.drop('dep_delayed_15min',axis=1,inplace=True)
```

```python
plt.figure(dpi=150)
sns.barplot(data=last_df,x='DayOfWeek',y='dep_delayed_15min_sum',order=['lunes','martes','miercoles','jueves','viernes','sabado','domingo'],color='purple')
plt.xlabel("Dia de la semana")
plt.ylabel("Cantidad")
plt.title("Cantidad de vuelos con retraso segun dia de la semana")
plt.savefig("visu6.jpg")
files.download("visu6.jpg") 
```
